```python
#Loading Dataset
movies = pd.read_csv("/kaggle/input/imdb-dataset-of-top-1000-movies-and-tv-shows/imdb_top_1000.csv" , header = 0)
```

```python
movies.head(4)
```

```python
#Extarcting Important Columns

col_names = ['Poster_Link','Series_Title','Overview','Director','Star1','Star2','Star3']
```

```python
movies = movies[col_names]
```

```python
movies[col_names].head(4)
```

```python
# Extract values from 'Overview','Director','Star1', 'Star2', 'Star3' columns and create a list
movies['tag'] = movies[['Overview','Director','Star1', 'Star2', 'Star3']].values.tolist()
```

<h1>Tag Column Formatting<h1>

```python
movies['tag'] = movies['tag'].apply(lambda x : str(x))
```

```python
#Removing Special Charaters
movies['tag'] = movies['tag'].apply(lambda x: ''.join(filter(lambda y: y.isalpha() or y.isspace(), x)))
```

<h1>Converting whole text in lowercase<h1>

```python
movies['tag'] = movies['tag'].apply(lambda x: x.lower()) 
```

```python
movies['tag'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
new_df = movies[['Poster_Link','Series_Title','tag']]
```

```python
new_df.head(2)
```

```python
cv = CountVectorizer(max_features=1500 , stop_words= 'english')
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x: ' '.join(y.upper() for y in x))
```

```python
#Converting string into list of words
new_df['tag'] = new_df['tag'].apply(lambda x : x.split())
```

```python
#convering list of words into a string
new_df['tag'] = new_df['tag'].apply(lambda x : ' '.join(x))
```

```python
# Initialize the Porter Stemmer
porter = PorterStemmer()

# Function to apply stemming to a text
def stem_text(text):
    tokens = word_tokenize(text)
    stemmed_tokens = [porter.stem(token) for token in tokens]
    return ' '.join(stemmed_tokens)

# Apply stemming to the 'tag' column
new_df['tag'] = new_df['tag'].apply(stem_text)
```

```python
#Calculating cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Similartiy Metrics
similarity = cosine_similarity(vectors)
```

```python
#filtering index based on Series_Title
new_df[new_df['Series_Title'] == 'The Dark Knight'].index[0]
```

```python
#Extracting top-5 movies based on cosine-similarity
sorted(list(enumerate(similarity[0])),reverse = True , key = lambda x : x[1])[1:6]
```

```python
def recommand(movie):
    index_value = new_df[new_df['Series_Title'] == movie].index[0]
    recommended_movies = sorted(list(enumerate(similarity[index_value])),reverse = True , key = lambda x : x[1])[1:6] # Retruns list of tuples of top 5 movies
    print("Recommended Movies:")
    print("-------------------")
    for index in range(5):
        print(new_df.at[recommended_movies[index][0], 'Series_Title']) # Extracting Moving Names corresponding to index value
```

```python
recommand('The Shawshank Redemption')
```
